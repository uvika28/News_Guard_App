```python
from nbconvert import PythonExporter
import nbformat


with open("train_model.ipynb",'r',encoding = 'utf-8') as f:
    N=nbformat.read(f,as_version=4) 
    
    
    
(script,_) =   PythonExporter().from_notebook_node(N)

with open("train_model.py",'w') as script_file:
    script_file.write(script)
```

```python
from nbconvert import PythonExporter

exporter = PythonExporter()
script, _ = exporter.from_notebook_node(N)

with open("app.py", "w", encoding="utf-8") as f:
    f.write(script)
```
